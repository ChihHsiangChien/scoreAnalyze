```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.font_manager import *
```

```python
#定義中文，字型檔用  fc-list :lang=zh 查詢

myfont = FontProperties(fname=r'C:\Windows\Fonts\NotoSansCJKtc-Regular.otf')
#myfont = FontProperties(fname='/usr/share/fonts/opentype/noto/NotoSansCJK-Regular.ttc')
#解決負號'-'顯示為方塊的問題
#matplotlib.rcParams['axes.unicode_minus']=False

```

```python
#讀取csv檔
#os.chdir("/media/pancala/3T/試題分析/105")
#file='105A.csv'
#data = pd.read_csv(file)


files_list = ['A1/生物學生分析資料.csv',
              'A2/生物學生分析資料.csv',
              'A3/生物學生分析資料.csv',
              'B1/生物學生分析資料.csv',]

x_label_list = ['上一' ,
                '上二' ,
                '上三' ,
                '下一' ,
                '下二' ,
                '下三' ,
               ]
```

```python
df = pd.DataFrame(columns=['班級', '座號','姓名'])

df = pd.read_csv(files_list[0])

df['pr1'] = df['答對率'].rank(pct = True)
df= df.rename(columns={'問題注意係數': '粗心度1'})


for i in range(1,4):
    df_sub = pd.read_csv(files_list[i])
    
    df_sub['pr' + str(i+1) ] = df_sub['答對率'].rank(pct = True)
    df_sub = df_sub.rename(columns={'問題注意係數': '粗心度' + str(i+1)})
    df = pd.merge(df, df_sub, on=['班級', '座號','姓名'])

df
```

```python
list(df.columns)
```

```python
mode_list = ['pr','粗心度']
```

```python
test_num = len(files_list)

def plotScore(stuclass, mode):

    #篩選出特定班級
    filter = df["班級"] == stuclass
    new_data = df[filter]

    #計算班級人數
    studentNum = new_data.shape[0]

    #subplot的列數由總班級人數決定
    subplotRowNum = int(np.ceil(studentNum/5))
    #把每一個學生的資料取出

    #設定subplots的列數、欄數和大小
    fig, axarr = plt.subplots(nrows=subplotRowNum,
                            ncols=5,
                            sharex=False,
                            sharey=False,
                            figsize=(10,subplotRowNum*2+2))

    #設定主標題
    fig.suptitle(str(stuclass)+ "班" +"生物段考" + mode + "變化",fontproperties=myfont, size=20)  

    #各小圖之間的間隔設定
    fig.tight_layout(pad=6, w_pad=1, h_pad=3.0)

    #把每一個學生的資料取出
    for i in range(0,studentNum):  
        #座號
        num = new_data['座號'].iloc[i]

        #座號前面補零
        if num < 10:
            num = str(0)+str(num)

        #姓名    
        name = new_data['姓名'].iloc[i]
        
        #取出PR
        pr_list = []
        for j in range(1,test_num+1):
            pr_list.append( new_data[ mode + str(j) ].iloc[i] )
            pr_list[j-1] =  pr_list[j-1]  * 100


        x = np.arange(1, test_num+1)
        y = pr_list


        #畫出PR折線圖
        axarr[int(i/5),i%5].plot(x,y)

        #畫出格線
        axarr[int(i/5),i%5].grid()


        #限定y軸大小
        axarr[int(i/5),i%5].set_ylim([-5,105])

        #設定小圖的標題為姓名
        axarr[int(i/5),i%5].set_title( str(num) + name + mode , fontproperties = myfont,size=15)

        #X軸標示設定
        axarr[int(i/5),i%5].set_xticks(x)
        axarr[int(i/5),i%5].set_xticklabels( x_label_list[0:test_num] ,fontproperties=myfont)
        axarr[int(i/5),i%5].tick_params(labelsize=10)

        #y軸標示設定
        if mode == 'pr':
            axarr[int(i/5),i%5].set_yticks([20, 55, 70, 85, 100])
        else:
            axarr[int(i/5),i%5].set_yticks([0, 50, 75, 100])
        
        #plt.show()

    #班級前面補零
    if stuclass <10:
        stuclass=str(0)+str(stuclass)
    else:
        stuclass=str(stuclass)


    fig.savefig(stuclass + mode + ".png",dpi = 250)
        #plt.close()

```

```python
stuclass_list = [1,5,12,18]

for stuclass in stuclass_list:
    for mode in mode_list:
        plotScore(stuclass , mode)
```

```python
test_num = len(files_list)

def plotScore_2(stuclass):

    #篩選出特定班級
    filter = df["班級"] == stuclass
    new_data = df[filter]

    #計算班級人數
    studentNum = new_data.shape[0]
    
    #計算要幾頁才能放下全班，通常兩頁
    page_num = int(np.ceil(studentNum / 18))

    #subplot的列數由總班級人數決定
    subplotRowNum = 6
    

    
    for page in range(page_num):
        #把每一個學生的資料取出

        #設定subplots的列數、欄數和大小
        fig, axarr = plt.subplots(nrows=subplotRowNum,
                                ncols=6,
                                sharex=False,
                                sharey=False,
                                figsize=(10,subplotRowNum*2+2))

        #設定主標題
        fig.suptitle(str(stuclass)+ "班" +"生物段考變化",fontproperties=myfont, size=10)  

        #各小圖之間的間隔設定
        fig.tight_layout(pad=6, w_pad=1, h_pad=6.0)

        #把每一個學生的資料取出
        if page == 0 : 
            start_stuNum = 0
            end_stuNum = 18
        else:
            start_stuNum = 18
            end_stuNum = studentNum
            
            
        for i in range(start_stuNum, end_stuNum):  
            #座號
            num = new_data['座號'].iloc[i]

            #座號前面補零
            if num < 10:
                num = str(0)+str(num)

            #姓名    
            name = new_data['姓名'].iloc[i]

            #取出PR
            pr_list = []
            for j in range(1,test_num+1):
                pr_list.append( new_data[ mode_list[0] + str(j) ].iloc[i] )
                pr_list[j-1] =  pr_list[j-1]  * 100


            #取出粗心度    
            problem_list = []
            for j in range(1,test_num+1):
                problem_list.append( new_data[ mode_list[1] + str(j) ].iloc[i] )
                problem_list[j-1] =  problem_list[j-1]  * 100
            
            x = np.arange(1, test_num+1)
            y_pr = pr_list
            y_problem = problem_list



            for mode in mode_list:

                if mode == 'pr':
                    m = (i%18) * 2 
                    y = y_pr
                else:
                    m = (i%18) * 2 + 1
                    y = y_problem

                plot_serial = (int(m/6), m%6)                    
                #畫出折線圖
                axarr[plot_serial].plot(x , y)

                #畫出格線
                axarr[plot_serial].grid()


                #限定y軸大小
                axarr[plot_serial].set_ylim([-5,105])

                #設定小圖的標題為姓名
                axarr[plot_serial].set_title( str(num) + name + '\n' + mode , fontproperties = myfont,size=15)

                #X軸標示設定
                axarr[plot_serial].set_xticks(x)
                axarr[plot_serial].set_xticklabels( x_label_list[0:test_num] ,fontproperties=myfont)
                axarr[plot_serial].tick_params(labelsize=10)

                #y軸標示設定
                if mode == 'pr':
                    axarr[plot_serial].set_yticks([20, 55, 70, 85, 100])
                else:
                    axarr[plot_serial].set_yticks([0, 50, 75, 100])

                #plt.show()

        #班級前面補零
        if stuclass < 10:
            stuclass_str = str(0)+str(stuclass)
        else:
            stuclass_str = str(stuclass)


        fig.savefig('合併_' + stuclass_str + '_' + str(page) + ".png",dpi = 250)
        #plt.close()

        
```

```python
stuclass_list = [1,5,12,18]

for stuclass in stuclass_list:
    plotScore_2(stuclass)
```
