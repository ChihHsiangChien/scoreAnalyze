```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.font_manager import *
```

```python
#定義中文，字型檔用  fc-list :lang=zh 查詢

myfont = FontProperties(fname=r'C:\Windows\Fonts\NotoSansCJKtc-Regular.otf')
#myfont = FontProperties(fname='/usr/share/fonts/opentype/noto/NotoSansCJK-Regular.ttc')
#解決負號'-'顯示為方塊的問題
#matplotlib.rcParams['axes.unicode_minus']=False

```

```python
#讀取csv檔
#os.chdir("/media/pancala/3T/試題分析/105")
#file='105A.csv'
#data = pd.read_csv(file)


files_list = ['A1/生物學生分析資料.csv',
              'A2/生物學生分析資料.csv',
              'A3/生物學生分析資料.csv',
              'B1/生物學生分析資料.csv',]

df1 = pd.read_csv(files_list[0])
df2 = pd.read_csv(files_list[1])
df3 = pd.read_csv(files_list[2])
df4 = pd.read_csv(files_list[3])
```

```python
df1['rank1'] = df1['答對率'].rank(pct = True)
df1 = df1.rename(columns={'問題注意係數': 'problem1'})

df2['rank2'] = df2['答對率'].rank(pct = True)
df2 = df2.rename(columns={'問題注意係數': 'problem2'})

df3['rank3'] = df3['答對率'].rank(pct = True)
df3 = df3.rename(columns={'問題注意係數': 'problem3'})

df4['rank4'] = df4['答對率'].rank(pct = True)
df4 = df4.rename(columns={'問題注意係數': 'problem4'})

           
```

```python
df = pd.merge(df1, df2, on=['班級', '座號','姓名'])
df = pd.merge(df, df3, on=['班級', '座號','姓名'])
df = pd.merge(df, df4, on=['班級', '座號','姓名'])

df
```

```python
list(df.columns)
```

```python
def plotPR(stuclass):

    #篩選出特定班級
    filter = df["班級"] == stuclass
    new_data = df[filter]

    #計算班級人數
    studentNum = new_data.shape[0]

    #subplot的列數由總班級人數決定
    subplotRowNum = int(np.ceil(studentNum/5))
    #把每一個學生的資料取出

    #設定subplots的列數、欄數和大小
    fig, axarr = plt.subplots(nrows=subplotRowNum,
                            ncols=5,
                            sharex=False,
                            sharey=False,
                            figsize=(10,subplotRowNum*2+2))

    #設定主標題
    fig.suptitle(str(stuclass)+ "班" +"生物段考PR變化",fontproperties=myfont, size=20)  
    #plt.title(str(stuclass)+ "班" +"上學期生物段考PR變化",fontproperties=myfont, fontsize=10)

    #各小圖之間的間隔設定
    fig.tight_layout(pad=6, w_pad=1, h_pad=3.0)

    #把每一個學生的資料取出
    for i in range(0,studentNum):  
        #座號
        num = new_data['座號'].iloc[i]

        #座號前面補零
        if num < 10:
            num = str(0)+str(num)

        #姓名    
        name = new_data['姓名'].iloc[i]

        #A1的PR
        a1 = new_data['rank1'].iloc[i]
        a1 = float(a1)*100

        #A2的PR
        a2 = new_data['rank2'].iloc[i]    
        a2 = float(a2)*100

        #A3的PR
        a3 = new_data['rank3'].iloc[i]
        a3 = float(a3)*100

        #B1的PR
        b1 = new_data['rank4'].iloc[i]
        b1 = float(b1)*100

        x = [1,2,3,4]
        y = [a1,a2,a3,b1]


        #畫出PR折線圖
        axarr[int(i/5),i%5].plot(x,y)

        #畫出格線
        axarr[int(i/5),i%5].grid()


        #限定y軸大小
        axarr[int(i/5),i%5].set_ylim([-5,105])

        #設定小圖的標題為姓名
        axarr[int(i/5),i%5].set_title(str(num) + name,fontproperties=myfont,size=15)

        #X軸標示設定為一二三
        axarr[int(i/5),i%5].set_xticks((1,2,3,4))
        axarr[int(i/5),i%5].set_xticklabels( ('上一','上二','上三','下一'),fontproperties=myfont)
        axarr[int(i/5),i%5].tick_params(labelsize=10)

        #plt.show()

    #班級前面補零
    if stuclass <10:
        stuclass=str(0)+str(stuclass)
    else:
        stuclass=str(stuclass)


    fig.savefig(stuclass + ".png",dpi = 250)
        #plt.close()

```

```python
stuclass_list = [1,5,12,18]

for stuclass in stuclass_list:
    plotPR(stuclass)
```
